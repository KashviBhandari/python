```python
import mysql.connector
```

```python
connection  = mysql.connector.connect(
host= "localhost",
user = "root",
password="admin",
database= "cafe",
)
```

```python
connection.is_connected()
```

```python
cursor = connection.cursor()
```

```python
query ="""
create table  cafe_manu(item_id int  AUTO_INCREMENT PRIMARY KEY ,
item_name VARCHAR(50), Price decimal (6,2));
"""
```

```python
try:
     cursor.execute(query)
except Exception as e:
     print(e)
```
