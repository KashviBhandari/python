Even and Odd Number Counter


```python
numbers = []
even = 0
odd = 0

for  n in range(10):
   num = int(input("enter number" ))
   numbers.append(num)
for num in numbers:
   if num  % 2==0:
     even += 1
   else :
      odd +=1
      

print("numbers", numbers) 
print("even", even)
print("odd" , odd)
```

 Word Counter

```python
def word_counter(sentence):
    words = sentence.split()
    total_words =len(words)
    longest = max(words , key = len)
    
    print("no of words:",total_words)
    print("Longest word:", longest)

sentence = input("Enter a sentence: ")
word_counter(sentence)
```

 Count Duplicate Elements

```python
numbers = [10, 20, 10, 30, 40, 20, 50]

Duplicate = []

for num in numbers:
    if numbers.count(num) > 1 and num not in Duplicate :
        Duplicate.append(num)

print("Duplicate" , Duplicate)

duplicates_exist=  len("Duplicate")

print("Duplicate exist" , duplicates_exist)
```

. Character Frequency Challenge

```python
word = input("enter word:")
Frequency = {}
for char in word:
    Frequency[char] = Frequency.get(char , 0) + 1
print(Frequency)

mostchar = max(Frequency, key=Frequency.get)
count = Frequency[mostchar]

print(mostchar,":", count)
```
