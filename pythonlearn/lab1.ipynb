```python
#========================lab 3==============================
#To create a Python program that finds and prints all numbers within a specified range that are divisible by 3 and are multiples of 5. 

start_number =  int(input("Enter start number :"))
end_number = int(input("Enter end number : "))

if start_number > end_number :
    print("the start number must be less than the ending number")
else :
     print(f"numbers between {start_number} and {end_number} that are divisible by 3 and  multiples of 5:")
for num in range(start_number , end_number + 1):
     if num % 3==0 and num % 5 == 0:
          print(num)
```
