```python
import spacy

# Load a pre-trained model
nlp = spacy.load("en_core_web_sm")

# Process text
doc = nlp("""MANGALURU: The Karnataka Land Grabbing Prohibition Special Court has directed the Belthangady police inspector to issue notices to
           Dharmasthala Dharmadhikari and Rajya Sabha member D Veerendra Heggade with regard to the alleged encroachment of government lands at 
          two places in Belthangady taluk of Dakshina Kannada. 
          The court has issued two orders— first on February 19 and second on April 2, 2026.""")

# Iterate through entities
for ent in doc.ents:
    print(ent.text, ent.label_)
# Output: Apple (ORG), U.K. (GPE), $1 billion (MONEY)
```
