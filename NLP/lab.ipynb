```python

import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer 
import requests
```

```python
df= pd.read_csv("C:\\Users\\User\\Desktop\\python\\NLP\\Womens Clothing E-Commerce Reviews.csv")
df.isna()
df.dropna(inplace= True)
df.columns
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
sia.polarity_scores("This dress is amazing!")
```

```python

sia = SentimentIntensityAnalyzer() 
def get_sentiment(score): 
    score = sia.polarity_scores(str(score))['compound']
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"
 
# Apply sentiment analysis 
df['sentiment'] = df['Review Text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
df['sentiment'] = df['Review Text'].apply(get_sentiment) 
 
# Show results 
df[[ 'Review Text', 'sentiment']].head(10)
```

```python
for sentiment in ['Positive', 'Neutral', 'Negative']: 
 print(f"\n--- {sentiment} Reviews ---") 
print(df[df['sentiment'] == sentiment]['Review Text'].head(3).to_string(index=False)) 
```
