```python
import tensorflow 
from tensorflow.keras import layers ,Sequential
```

```python
train = tensorflow.keras.utils.image_dataset_from_directory("C:\\Users\\User\\Desktop\\python\\deeplearing\\archive (1)\\training_set",image_size = (150,150))
test= tensorflow.keras .utils.image_dataset_from_directory("C:\\Users\\User\\Desktop\\python\\deeplearing\\archive (1)\\test_set",image_size=(150,150))
```

```python
model =Sequential([
    layers.Rescaling(1./255,input_shape=(150,150,3)),
    layers.Conv2D(64,(3,3),activation= "relu"),
    layers.MaxPool2D((2,2)),

 layers.Flatten(),
 layers.Dense(16, activation="relu"),
 layers.Dense(8,activation="relu"),
 layers.Dense(1,activation= "sigmoid"),
])
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam", metrics=["accuracy"])
model.summary()
```

```python
model.fit(train,
          validation_data= test,
          epochs= 10,
          batch_size = 64)
```

```python
model.predict(test)
```
